# Code Block 7: Data Preprocessing and Missing Data Check for 'Region' (Categorical)
This block visualizes the 'Region' column, which is categorical:
1. **Bar Chart**: Shows the distribution of countries across different regions. It helps identify how many countries belong to each region, aiding in a better understanding of the regional representation in the dataset.
2. **Pie Chart**: Displays the proportion of countries in each region as percentages, offering a clear breakdown of the dataset's regional composition.
