```python
import pdfplumber

# Path to the new PDF file
pdf_path_new = r"C:\Users\jamar\Downloads\Topic_4_Data_Visualization_and_Charts_Data.pdf"

# Extract text from the new PDF
data_new = {}
with pdfplumber.open(pdf_path_new) as pdf:
    for page_num, page in enumerate(pdf.pages):
        text = page.extract_text()
        data_new[page_num] = text

# Print the extracted text from the first page
print(data_new[0])  # Show text from the first page as an example
```

This code creates a bar chart that displays the percentage of favorable opinions for each country. The countries are listed on the x-axis, and the percentage of favorable opinions is represented by the height of the bars. The color of the bars is set to green.

```python
import matplotlib.pyplot as plt
plt.bar(['Canada', 'U.S.', 'Poland', 'Sweden', 'Philippines', 'Brazil', 'Nigeria', 'Japan'], [63, 52, 81, 76, 83, 51, 69, 41], color='green')
plt.title('Favorable Opinion by Country'); plt.show()
```

This code generates a horizontal bar chart showing the percentage of unfavorable opinions for each country. The countries are displayed on the y-axis, and the percentages are represented by the length of the bars. The bars are colored red.

```python
plt.barh(['Canada', 'U.S.', 'Poland', 'Sweden', 'Philippines', 'Brazil', 'Nigeria', 'Japan'], [31, 44, 10, 18, 14, 30, 22, 50], color='red')
plt.title('Unfavorable Opinion by Country'); plt.show()
```

This code creates a pie chart that displays the distribution of "Don't Know/Refusal" responses for each country. Each slice represents a country, with its size corresponding to the percentage of DK/Refusal. The autopct parameter displays the percentage on each slice, and startangle rotates the chart for better appearance.

```python
plt.pie([6, 4, 9, 6, 3, 19, 9, 9], labels=['Canada', 'U.S.', 'Poland', 'Sweden', 'Philippines', 'Brazil', 'Nigeria', 'Japan'], autopct='%1.1f%%', startangle=140)
plt.title('DK/Refusal Distribution by Country'); plt.axis('equal'); plt.show()
```

This code produces a line plot comparing favorable and unfavorable opinions across countries. Two lines are drawn: one for favorable opinions (green) and one for unfavorable opinions (red), with markers indicating the data points. The legend distinguishes between the two lines.

```python
plt.plot(['Canada', 'U.S.', 'Poland', 'Sweden', 'Philippines', 'Brazil', 'Nigeria', 'Japan'], [63, 52, 81, 76, 83, 51, 69, 41], marker='o', label='Favorable', color='green')
plt.plot(['Canada', 'U.S.', 'Poland', 'Sweden', 'Philippines', 'Brazil', 'Nigeria', 'Japan'], [31, 44, 10, 18, 14, 30, 22, 50], marker='o', label='Unfavorable', color='red'); plt.title('Favorable vs Unfavorable Opinion'); plt.legend(); plt.show()
```

This code creates a box plot that summarizes the distribution of favorable and unfavorable opinions. The box plot provides insights into the median, quartiles, and potential outliers for each category. The x-axis labels distinguish between favorable and unfavorable opinions.

```python
import seaborn as sns
sns.boxplot(data=[[63, 52, 81, 76, 83, 51, 69, 41], [31, 44, 10, 18, 14, 30, 22, 50]])
plt.title('Box Plot of Opinions'); plt.xticks(ticks=[0, 1], labels=['Favorable', 'Unfavorable']); plt.show()
```

This code generates a scatter plot that visualizes the relationship between favorable and unfavorable opinions. Each point represents a country, plotted with favorable opinions on the x-axis and unfavorable opinions on the y-axis. The grid helps in interpreting the data points.

```python
plt.scatter([63, 52, 81, 76, 83, 51, 69, 41], [31, 44, 10, 18, 14, 30, 22, 50], color='purple')
plt.title('Scatter Plot of Opinions'); plt.xlabel('Favorable (%)'); plt.ylabel('Unfavorable (%)'); plt.grid(); plt.show()
```

This code creates an area plot that visually represents the favorable opinions for each country. The area under the line is filled with orange, helping to emphasize the total proportion of favorable opinions across countries.

```python
plt.fill_between(['Canada', 'U.S.', 'Poland', 'Sweden', 'Philippines', 'Brazil', 'Nigeria', 'Japan'], [63, 52, 81, 76, 83, 51, 69, 41], color='orange', alpha=0.5)
plt.title('Favorable Opinion Area Plot'); plt.show()
```

This code generates a heatmap that visually represents the favorable and unfavorable opinions for each country. The cells are color-coded, with the annotations showing the exact percentages. The heatmap provides a quick overview of the opinion data across countries.

```python
sns.heatmap([[63, 52, 81, 76, 83, 51, 69, 41], [31, 44, 10, 18, 14, 30, 22, 50]], annot=True, xticklabels=['Canada', 'U.S.', 'Poland', 'Sweden', 'Philippines', 'Brazil', 'Nigeria', 'Japan'], yticklabels=['Favorable', 'Unfavorable'])
plt.title('Heatmap of Opinions'); plt.show()
```
